## Birleştirme -- Join -- İşlemleri

```python
import numpy as np
import pandas as pd

m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns= ["var1","var2","var3"])
df1
```

```python
df2 = df1 + 99 
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
# index problemi ortadan kalkar. ve dfler birleşir.
pd.concat([df1,df2], ignore_index=True)
```

```python
df1.columns
```

```python
df2.columns
```

```python
df2.columns = ["var1","var2","deg3"]
```

```python
df2
```

```python
# değişken değerleri örtüşmüyorsa hatalı şekilde dönecek..
pd.concat([df1,df2])
```

```python
## kesisimlerini birleştirir. join="inner"
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], join_axes = [df1.columns] , ignore_index=True)
```
